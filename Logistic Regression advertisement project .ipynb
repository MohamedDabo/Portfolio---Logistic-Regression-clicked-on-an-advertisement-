# Logistic Regression

Now it's time to do a train test split, and train our model!

You'll have the freedom here to choose columns that you want to train on!

** Split the data into training set and testing set using train_test_split**

```python
model_data = ad_data.drop(['Country','City','Ad Topic Line','Timestamp'],axis=1)
model_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = model_data.drop('Clicked on Ad',axis=1)
y = model_data['Clicked on Ad']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

** Train and fit a logistic regression model on the training set.**

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmod = LogisticRegression()
logmod.fit(X_train,y_train)
```

## Predictions and Evaluations
** Now predict values for the testing data.**

```python
predictions = logmod.predict(X_test)
```

** Create a classification report for the model.**

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print('Confusion Matrix:')
print(confusion_matrix(y_test,predictions))
print('\nClassification Report:')
print(classification_report(y_test,predictions))
```

Our model has reached a precision of 92%. With only 8 false positives and 15 false negatives, this is a pretty satisfying result.
